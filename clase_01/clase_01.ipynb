### 🎢 **`Motivación: Reproducibilidad y Riesgo de Modelo en Finanzas`**

#### ⛱️ ¿Por qué empezar por aquí?

La mayoría de los cursos de ML empiezan con algoritmos. Este curso empieza con **infraestructura** porque en finanzas las consecuencias de un modelo irreproducible son graves:

- **Regulatorias**: el Banco Central, la CMF o la Fed pueden requerir que un banco reproduzca exactamente los resultados de un modelo de crédito de hace 2 años
  
- **Económicas**: un backtest brillante que no se puede replicar en producción cuesta millones
  
- **Reputacionales**: modelos que cambian de resultado según quién los corra destruyen la confianza interna

#### 🛡️ El marco regulatorio que importa: SR 11-7

La **Supervisory Guidance on Model Risk Management** (especialmente conocida como SR 11-7 en EE.UU.) es un conjunto de directrices emitidas por reguladores bancarios (Fed, OCC) que establece estándares estrictos para identificar, medir y controlar el riesgo derivado del uso de modelos cuantitativos en instituciones financieras.

Ahora, ¿cómo se define un modelo bajo SR 11-7?

<tr>
<td><img src="../figs/figura_001.png" align="middle" style="width: 1000px;"/> </td>
</tr>

La SR 11-7, establece que los modelos financieros deben:

* Estar documentados completamente
* Ser validados por partes independientes
* Ser monitoreados en producción (model monitoring)
* Tener un proceso de gobernanza claro (model inventory, ownership)
* Poder ser reproducidos y auditados


Esta guía, aunque es de regulación bancaria en EE.UU., es el estándar de facto para la industria financiera global. En Chile, la CMF avanza hacia lineamientos similares.

**Lo que esto significa para nosotros**: cada notebook, cada modelo, cada resultado que produzcamos en este curso debe poder ser reproducido por cualquier otra persona con acceso al repositorio. Eso **no es opcional** en la industria.

#### 🪁 Tres casos reales de falla

##### Caso 1 — El modelo de crédito que nadie puede replicar

> Un banco desarrolla un modelo de scoring crediticio en 2022. En 2024 hay una auditoría interna. El data scientist que lo construyó ya no está en la empresa. El notebook tiene celdas ejecutadas fuera de orden, las versiones de librerías no están fijadas, los datos de entrenamiento fueron sobreescritos. **Resultado**: el banco no puede demostrar que el modelo fue construido correctamente. Exposición regulatoria significativa.

##### Caso 2 — El backtest que no se puede replicar en producción

> Un quant desarrolla una estrategia de trading algorítmico con Sharpe Ratio de 2.3. Cuando el equipo de riesgo intenta replicarlo, el Sharpe cae a 0.4. La causa: **data leakage** — la función de features usaba información del futuro sin querer. Sin un pipeline reproducible y controlado por versiones, el error tomó semanas en encontrarse.

##### Caso 3 — El modelo de riesgo con dependencia de sistema operativo

> Un modelo de VaR (Value at Risk) da resultados ligeramente distintos en Windows vs Linux debido a diferencias en la implementación de floats en ciertas librerías. Sin un entorno reproducible, esto se descubre solo cuando los resultados de dos analistas no cuadran. **Resultado**: horas de debugging, desconfianza en el modelo.

---

**La solución a los tres escenarios es la misma**: estructura de proyecto clara + control de versiones con Git + entorno reproducible.

###  🎉 **`Estructura de Proyectos DS/ML`**

#### 🦜 El problema de los proyectos sin estructura

Un proyecto típico de ML sin estructura suele verse así:

```
proyecto/
├── modelo_final.ipynb
├── modelo_final_v2.ipynb
├── modelo_final_v2_ESTE.ipynb
├── modelo_final_v2_ESTE_corregido.ipynb
├── datos.csv
├── datos_nuevo.csv
├── datos_nuevo_limpio.csv
└── resultados_buenos.xlsx
```

Esto, en la práctica, es inauditable, irreproducible y no escalable (salvo para fines pedag´gicos)

#### 🍁 La estructura estándar para este curso

Todos los trabajos y proyectos del curso seguirán esta estructura:

```
proyecto_finanzas/
│
├── data/
│   ├── raw/              ← Datos originales. NUNCA se modifican.
│   ├── processed/        ← Features, datasets limpios, transformados
│   └── external/         ← Benchmarks, índices de referencia, datos macro
│
├── notebooks/
│   ├── 01_eda.ipynb      ← Exploración: no va a producción
│   ├── 02_features.ipynb ← Desarrollo de features
│   └── 03_modeling.ipynb ← Experimentos de modelado
│
├── src/                  ← Código modular, funciones reutilizables
│   ├── __init__.py
│   ├── data.py           ← Carga y validación de datos
│   ├── features.py       ← Ingeniería de variables
│   ├── validation.py     ← Walk-forward CV, time series split
│   └── metrics.py        ← Métricas financieras (Sharpe, MDD, etc.)
│
├── models/               ← Artefactos serializados + metadata de cada modelo
│   └── baseline_v1/
│       ├── model.pkl
│       └── metadata.json
│
├── reports/              ← Outputs finales: gráficos, tablas, reportes
│   └── figures/
│
├── tests/                ← Tests unitarios para funciones en src/
│
├── .gitignore            ← Excluye datos sensibles, credenciales, binarios
├── environment.yml       ← Dependencias del entorno (reproducibilidad)
└── README.md             ← Descripción, instrucciones, resultados clave
```

#### 🍂 Principios detrás de la estructura

| Principio | Aplicación |
|---|---|
| **Inmutabilidad de datos raw** | `data/raw/` nunca se modifica. Si necesitas limpiar, crea un script que genere `data/processed/` |
| **Separación exploración/producción** | Notebooks para explorar, `src/` para código que va a producción |
| **Trazabilidad de modelos** | Cada modelo en `models/` tiene su `metadata.json` con parámetros, fecha, métricas |
| **Secretos fuera del repo** | API keys, passwords → variables de entorno o `.env` (en `.gitignore`) |
| **README como contrato** | Cualquier persona debe poder reproducir el proyecto solo con el README |

````python
# Creación programática de la estructura de proyecto
# Ejecutar esto para inicializar cualquier nuevo proyecto del curso

from pathlib import Path
from datetime import date
import json

def crear_estructura_proyecto(nombre_proyecto: str, base_dir: str = ".") -> None:
    """
    Crea la estructura estándar de directorios para un proyecto ML Finance.

    Parameters
    ----------
    nombre_proyecto : str
        Nombre del proyecto (se convierte a snake_case)
    base_dir : str
        Directorio base donde crear el proyecto
    """
    nombre = nombre_proyecto.lower().replace(" ", "_")
    raiz = Path(base_dir) / nombre

    directorios = [
        "data/raw", "data/processed", "data/external",
        "notebooks", "src", "models", "reports/figures", "tests",
    ]

    for d in directorios:
        (raiz / d).mkdir(parents=True, exist_ok=True)
        (raiz / d / ".gitkeep").touch()

    (raiz / "src" / "__init__.py").write_text("")

    gitignore = """# Datos y modelos grandes
data/raw/*
data/processed/*
!data/raw/.gitkeep
!data/processed/.gitkeep
models/**/*.pkl
models/**/*.joblib
models/**/*.h5

# Credenciales y secretos
.env
*.key
credentials.json
secrets.yaml

# Python
__pycache__/
*.pyc
*.pyo
.ipynb_checkpoints/

# Entornos
.venv/
venv/

# Sistema operativo
.DS_Store
Thumbs.db

# IDEs
.vscode/
.idea/

# Outputs temporales
*.log
mlruns/"""

    (raiz / ".gitignore").write_text(gitignore)

    readme = f"""# {nombre_proyecto}

**Creado**: {date.today().isoformat()}  
**Curso**: Machine Learning Aplicado a las Finanzas — USACH

## Descripción

> Describir aquí el problema financiero que aborda el proyecto.

## Reproducción

```bash
git clone <url>
cd {nombre}
conda env create -f environment.yml
conda activate {nombre}
jupyter lab
```

## Estructura

```
data/       → Datos (raw inmutable, processed derivado)
notebooks/  → Exploración y desarrollo
src/        → Código modular reutilizable
models/     → Artefactos serializados + metadata
reports/    → Resultados finales
```
"""
    (raiz / "README.md").write_text(readme)

    env_yml = f"""name: {nombre}
channels:
  - conda-forge
  - defaults
dependencies:
  - python=3.11
  - pip
  - pandas>=2.0
  - numpy>=1.25
  - scikit-learn>=1.3
  - matplotlib>=3.7
  - seaborn>=0.12
  - jupyterlab>=4.0
  - ipykernel
  - pip:
    - yfinance>=0.2.30
    - xgboost>=2.0
    - lightgbm>=4.0
    - shap>=0.44
    - mlflow>=2.10
    - python-dotenv>=1.0
"""
    (raiz / "environment.yml").write_text(env_yml)

    print(f"Proyecto '{nombre}' creado en: {raiz.resolve()}")
    print("\nEstructura generada:")
    for item in sorted(raiz.rglob("*")):
        nivel = len(item.relative_to(raiz).parts) - 1
        prefijo = "    " * nivel + ("├── " if nivel > 0 else "")
        print(f"{prefijo}{item.name}")


# Crear un proyecto de práctica
crear_estructura_proyecto("practica_setup", base_dir="/tmp")
````

#### 📌 ¿Qué es Git?

Es el sistema de control de versiones en si mismo. Es una herramienta de linea de comandos que se instala localmente en el computador.

* Fue creado por Linus Torvalds en el 2005 (el mismo creado de linux)
* Registra cada cambio en el codigo con autor, fecha y mensaje
* Es el standar de la industria de desarrollo de software

#### 📍 ¿Qué es Github?

Es una plataforma web que aloja repositorios Git en la nube.

* Propiedad de Microsoft (desde el 2018)
* Es el repositorio mas popular del mundo (~1000M de usuarios)
* Gratuito para proyectos publicos y privados (con limites)
* Funcionalidades destacadas: pull requests, issues, Github-Actions (CI/CD), Github Copilot.

#### 🧵 ¿Qué es Gitlab?

Es una alternativa a Github, tambien es una plataforma web para almacenar repositorios Git.

* Empresa Independiente (GitLab Inc.)
* Mas orientado a Devops Empresarial
* Su diferencial: Incluye Pipeline CI/CD, gestion de proyectos, monitoreo - todo integrado. 
* Permite instalacion on-premise (en tus propios servidores), lo que lo hace popular en empresas con requisitos de privacidad.

#### 🧶 ¿Cómo se instala Git?

1. Descargar `git` desde https://github.com/git-guides/install-git
2. Crearse una cuenta en `Github` en https://github.com/

#### 🎾 ¿Por qué Git y no solo "guardar el archivo"?

Git no es solo backup. Para proyectos ML en finanzas, Git provee:

| Capacidad | Valor en finanzas |
|---|---|
| **Historial completo** | Puedo reconstruir exactamente qué modelo existía el 15 de marzo de 2024 |
| **Branches** | Experimento con un nuevo conjunto de features sin romper el modelo en producción |
| **Tags** | Marco versiones de modelos: `v1.0-baseline`, `v2.1-con-macro-features` |
| **Blame/Log** | Sé quién cambió qué y cuándo — crucial para auditorías |
| **Revert** | Si el modelo en producción empeora, vuelvo al commit anterior en minutos |

#### 🏓 Convenciones de commits: Conventional Commits

Adoptamos el estándar [Conventional Commits](https://www.conventionalcommits.org/) adaptado para ML:

```bash
# Formato: <tipo>(<scope>): <descripción corta>

# Tipos para proyectos ML Finance:
feat:    # nueva funcionalidad o feature de datos
fix:     # corrección de bug (especialmente: data leakage, errores de validación)
exp:     # experimento (prueba un modelo, hiperparámetro, configuración)
data:    # cambio en datos o pipeline de datos
refac:   # refactorización sin cambio de comportamiento
docs:    # documentación
test:    # tests unitarios

# Ejemplos de como se usan en la practica:
git commit -m "feat(features): agrega rolling_std_20d y RSI como predictores"
git commit -m "exp(model): prueba LightGBM lr=0.01 — AUC 0.74 en OOT"
git commit -m "fix(validation): corrige data leakage en split train/test"
git commit -m "data(raw): agrega datos IPSA 2015-2024 desde Yahoo Finance"
```

> El mensaje de commit es **el primer nivel de documentación de un experimento**. Un buen historial de Git es un log de experimentos legible.

#### 🚧 Branch (Ramas)

En machine learning, un error en producción no es solo un bug — puede traducirse en pérdidas monetarias, decisiones de inversión erróneas o reportes regulatorios incorrectos. Las ramas (branches) van a ser nuestra red de seguridad.

<tr>
<td><img src="../figs/figura_006.png" align="middle" style="width: 1000px;"/> </td>
</tr>

* **Razones especificas al contexto** Cuando estás iterando sobre un modelo predictivo (ej. predecir retornos, estimar riesgo), necesitas probar variantes:
  
  * `feature/lstm-prediction-precio`
  * `feature/xgboost-credit-scoring`
  * `feature/new-features`
  
* **Reproducibilidad** Los reguladores financieros (CMF, SEC, Basel III) y auditorías internas exigen que puedas reproducir exactamente cómo se tomó una decisión en una fecha específica. Con ramas + tags puedes reconstruir el estado exacto del modelo que generó un reporte.
  
* **Separacion de Entornos** (experimentacion vs produccion) 
  * `main` o `master` -> Modelo en produccion (stable)
  * `develop` -> Integracion de nuevas features
  * `feature` -> experimentacion individual
  * `hotfix` -> correciones urgentes 
  
* **Colaboracion con trazabilidad** permite auditar el modelo de punta a punta para ver qué cambio, quién lo modificó y por qué se cambio.

#### 🚫 Lo que NO va en Git


* Archivos de datos grandes (>50MB) → usar DVC o almacenamiento externo
* Credenciales, API keys, passwords → variables de entorno / .env
* Modelos serializados grandes (.pkl, .h5) → DVC o MLflow artifacts
* Datos de clientes o información sensible → nunca en repositorio, ni privado


### 🔨 **`Entornos Python Reproducibles`**

#### 🦺 El problema de "funciona en mi máquina"

Un modelo de ML (en general) puede dar resultados distintos entre dos máquinas si las versiones de librerías difieren. Esto no es teórico:

- `scikit-learn 1.2` vs `1.4` puede tener diferencias en la implementación de ciertos solvers
- `numpy` tiene cambios de comportamiento entre versiones menores en operaciones de álgebra lineal
- `pandas` cambia APIs entre versiones (ej: `DataFrame.append` deprecado en 1.4)

**La solución**: fijar las versiones de todas las dependencias en un archivo que sea parte del repositorio.

#### ⛑️ Opciones de gestión de entornos

| Herramienta | Pros | Contras | Cuándo usar |
|---|---|---|---|
| `conda` + `environment.yml` | Gestiona Python + librerías no-Python | Más pesado | Proyectos de data science con dependencias científicas |
| `pip` + `requirements.txt` | Simple, universal | No gestiona Python mismo | Proyectos simples o entornos controlados |
| `poetry` + `pyproject.toml` | Resolución de dependencias robusta | Curva de aprendizaje | Desarrollo de librerías o aplicaciones |
| `uv` | Ultrarrápido, moderno | Ecosistema nuevo | Proyectos nuevos donde la velocidad importa |

#### 🍪 ¿Por qué MLOps es especialmente crítico en finanzas?

| Razón | Impacto |
|---|---|
| **Data drift** | El mercado cambia de régimen. Un modelo pre-COVID puede fallar post-COVID |
| **Regulación** | SR 11-7, BCBS 239 requieren documentación y auditabilidad de modelos |

| **Tiempo real** | Modelos de trading o riesgo que fallan en producción tienen impacto inmediato en P&L |

#### 🍪 Etapas de MLOps

Las etapas van desde conseguir buenos datos hasta asegurarse de que el modelo funcione bien en el tiempo, de forma automatizada, confiable y bajo control.

<tr>
<td><img src="../figs/figura_005.png" align="middle" style="width: 1000px;"/> </td>
</tr>

* **Gestion de Datos** Los modelos de ML son tan buenos como los datos que los alimentan. Esta etapa se encarga de recolectar, limpiar y versionar los datos. Sin datos confiables, cualquier modelo fallará independiente de su sofisticación. Es el cimiento de todo el proceso.

* **Experimentacion**  Es la etapa científica del proceso. El equipo de data science prueba distintos enfoques, algoritmos e ideas para resolver el problema. Se registran todos los experimentos (métricas, parámetros, resultados) para poder comparar y reproducir los mejores resultados.

* **Desarrollo del Modelo** Una vez identificado el mejor enfoque, se construye el modelo definitivo: se entrena con más datos, se afinan sus parámetros y se valida rigurosamente. Al finalizar, el modelo queda empaquetado y registrado, listo para salir del laboratorio.

* **CI/CD (Integración y Entrega Continua)** Es el puente entre el desarrollo y la producción. Automatiza las pruebas del código y del modelo para garantizar que cada cambio no rompa nada. Si todo pasa los tests, el pipeline lo lleva automáticamente a producción, reduciendo errores humanos.

* **Despliegue** El modelo sale al mundo real. Se empaqueta como un servicio (API) que otros sistemas pueden consultar. Se elige cómo liberar la nueva versión: de golpe, gradualmente o en paralelo, minimizando el riesgo de afectar a los usuarios finales.

* **Monitoreo** Un modelo en producción puede degradarse silenciosamente. Los datos del mundo real cambian con el tiempo y el modelo puede volverse impreciso sin que nadie lo note. Esta etapa vigila continuamente el comportamiento del modelo y activa alertas o reentrenamientos cuando es necesario.

* **Gobierno y Compliance** Responde la pregunta: ¿podemos confiar y justificar las decisiones del modelo? Garantiza que el modelo sea reproducible, explicable, seguro y cumpla con regulaciones legales. En finanzas esto es crítico, ya que los reguladores exigen transparencia en decisiones automatizadas.
